```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
query_image=cv2.imread('queryimage.png')
ref1_image=cv2.imread('refimage1.png')
ref2_image=cv2.imread('refimage2.png')

query_image=cv2.cvtColor(query_image,cv2.COLOR_BGR2RGB)
plt.imshow(query_image)
plt.title("query image")
plt.show()

ref1_image=cv2.cvtColor(ref1_image,cv2.COLOR_BGR2RGB)
plt.imshow(ref1_image)
plt.title("ref1 image")
plt.show()

ref2_image=cv2.cvtColor(ref2_image,cv2.COLOR_BGR2RGB)
plt.imshow(ref2_image)
plt.title("ref2 image")
plt.show()

```

```python
plt.hist(query_image.ravel(),256,[0,256])
plt.title("query image")
plt.show()

plt.hist(ref1_image.ravel(),256,[0,256])
plt.title("ref1 image")
plt.show()

plt.hist(ref2_image.ravel(),256,[0,256])
plt.title("ref2 image")
plt.show()

```

```python
plt.figure(figsize=(20,5))

query_hist=cv2.calcHist([query_image], [0, 1, 2], None, [8, 8, 8], [0, 256, 0, 256, 0, 256])
query_hist_norm=cv2.normalize(query_hist,query_hist).flatten()
plt.subplot(1,3,1)
plt.plot(query_hist_norm)


ref1_hist=cv2.calcHist([ref1_image], [0, 1, 2], None, [8, 8, 8], [0, 256, 0, 256, 0, 256])
ref1_hist_norm=cv2.normalize(ref1_hist,query_hist).flatten()
plt.subplot(1,3,2)
plt.plot(ref1_hist_norm)

ref2_hist=cv2.calcHist([ref2_image], [0, 1, 2], None, [8, 8, 8], [0, 256, 0, 256, 0, 256])
ref2_hist_norm=cv2.normalize(ref2_hist,query_hist).flatten()
plt.subplot(1,3,3)
plt.plot(ref2_hist_norm)







```

```python
dist1=cv2.compareHist(query_hist_norm,ref1_hist_norm,cv2.HISTCMP_BHATTACHARYYA)
dist2=cv2.compareHist(query_hist_norm,ref2_hist_norm,cv2.HISTCMP_BHATTACHARYYA)
print(dist1)
print(dist2)
```
