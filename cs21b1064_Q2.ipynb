```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
def calculatecorr(data) :
    m = data.shape[1]
    n = data.columns
    print(n)
    correlated = np.zeros(m**2).reshape([m,m])
    for i in range(m) :
        for j in range(m) :
            if i==j :
                correlated[i][j] = 1
            else :
                mean1 = data[n[i]].mean()
                mean2 =data[n[j]].mean()
                X = data[n[i]] - mean1
                print(X)
                Y = data[n[j]] - mean2
                correlated[i][j] = ((X*Y).sum())/np.sqrt(((X**2).sum())*((Y**2).sum()))    
    return correlated
```

```python
data = pd.read_csv('iris.csv')
data.drop(labels='Species',axis=1,inplace = True)
```

```python
corr = calculatecorr(data)
print(corr)
```

```python
sns.heatmap(data.corr())
```
